```python
from matplotlib import pyplot as plt
```

```python
X = [0, 3, 6, 9]
Y = [20.3, 22, 24]
```

```python
Z_purity = [[0.996, 0.981, 0.982],
            [0.999, 0.995, 1.000]]
```

```python
plt.rcParams.update({'font.size': 18})
plt.pcolormesh(X, Y, Z_purity, cmap='bwr', vmax=1.0, vmin=0.95)

plt.xticks([3,6])
plt.yticks([20.3, 22])
plt.xlabel('S/N')
plt.ylabel(r'log $N_{\mathrm{HI}}$')
plt.title('Purity')
plt.colorbar()
```

```python
Z_completeness = [[0.962, 0.994, 0.993],
                  [0.960, 0.990, 1.000]]
```

```python
plt.rcParams.update({'font.size': 18})
plt.pcolormesh(X, Y, Z_completeness, cmap='bwr', vmax=1.0, vmin=0.95)

plt.xticks([3,6])
plt.yticks([20.3, 22])
plt.xlabel('S/N')
plt.ylabel(r'log $N_{\mathrm{HI}}$')
plt.title('Completeness')
plt.colorbar()
```
